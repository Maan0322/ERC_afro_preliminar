```python
import re

def encontrar_municipio(texto):

    if pd.isna(texto):
        return None

    texto = str(texto).upper()

    for m in municipios:
        if m in texto:
            return m

    return None


def encontrar_departamento(texto):

    if pd.isna(texto):
        return None

    texto = str(texto).upper()

    for d in departamentos:
        if d in texto:
            return d

    return None
```

```python
import pandas as pd

def obtener_todos_los_duplicados(df, cols_comparar):
    # Convertimos a lista para evitar el KeyError
    lista_cols = list(cols_comparar)
    # keep=False marca todas las ocurrencias de los duplicados
    return df[df.duplicated(subset=lista_cols, keep=False)].copy()

def contar_cruces_entre_bases(df_dups, base_a, base_b, cols_comparar):
    # Convertimos a lista para que pd.merge reconozca las múltiples columnas
    lista_cols = list(cols_comparar)
    
    # Filtramos por base y eliminamos duplicados internos de CADA base 
    # para que el cruce sea "Cuántos registros de A existen en B"
    df_a = df_dups[df_dups['base_de_datos'] == base_a].drop(columns=['base_de_datos']).drop_duplicates()
    df_b = df_dups[df_dups['base_de_datos'] == base_b].drop(columns=['base_de_datos']).drop_duplicates()
    
    # El merge ahora recibirá una lista real de strings
    coincidencias = pd.merge(df_a, df_b, on=lista_cols, how='inner')
    
    return len(coincidencias)

# --- USO DE LAS FUNCIONES CORREGIDO ---

# 1. Definir columnas de comparación como LISTA
cols = df_union.columns.drop("base_de_datos").tolist() 

# 2. Extraer el universo de los 943 duplicados
df_duplicados_total = obtener_todos_los_duplicados(df_union, cols)

# 3. Identificar cruces específicos (ahora sin errores de llave)
cruce_1_2 = contar_cruces_entre_bases(df_duplicados_total, 1, 2, cols)
cruce_0_1 = contar_cruces_entre_bases(df_duplicados_total, 0, 1, cols)
cruce_0_2 = contar_cruces_entre_bases(df_duplicados_total, 0, 2, cols)
# Añadimos el cruce con la base 4 que mencionaste antes
cruce_0_4 = contar_cruces_entre_bases(df_duplicados_total, 0, 4, cols)

print(f"Total registros duplicados en el set (filas totales): {len(df_duplicados_total)}")
print("-" * 40)
print(f"Registros únicos de Base 0 que también aparecen en Base 1: {cruce_0_1}")
print(f"Registros únicos de Base 0 que también aparecen en Base 2: {cruce_0_2}")
print(f"Registros únicos de Base 1 que también aparecen en Base 2: {cruce_1_2}")
print(f"Registros únicos de Base 0 que también aparecen en Base 4: {cruce_0_4}")
```
